## Data Preprocessing

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
# Load initial data
initial_data = pd.read_csv("metherology_dataset.csv")
initial_data.columns = initial_data.columns.str.strip()
```

### Understanding the Data

```python
# Display types of the columns~
print("Data types of each column:")
print(initial_data.dtypes)
# Number of values in the dataset
print("\nNumber of values in the dataset:")
print(initial_data.count())
# Check for missing values
print("\nMissing values per column:")
print(initial_data.isnull().sum())
```

### Checking for Errors and Outliers

```python
# Duplicates Check
print("\nNumber of duplicate rows in the dataset:")
print(initial_data.duplicated().sum())
```
